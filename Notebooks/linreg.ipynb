```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pathlib
import statsmodels.api as sm
```

```python
path = pathlib.Path().cwd().parents[0] / 'CSVs' / 'CarInsurance.csv'
df = pd.read_csv(path)
df.head()
```

```python
X = sm.add_constant(pd.DataFrame(df['Age']))
Y = pd.DataFrame(df['Annual_Premium'])
```

```python
model = sm.OLS(Y, X).fit()

display(model.summary())

print(f'The linear regression model is: \n Y = {model.params[0]:.3f} + {model.params[1]:.3f} * X')
```

The coefficient $73.451$ means that the `Annual_Premium` increases by $\$73.451$ for each additional year of age.
